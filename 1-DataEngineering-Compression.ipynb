# Data Engineering

## Turn all files into parquet for compression




```python
#create Spark session
spark = SparkSession.builder.appName('turntoparquet').getOrCreate()
```

```python
#turn chicago communities csv into parquet

chi_communities = spark.read \
    .option("quote", "\"")  \
    .option("escape", "\"") \
    .option("ignoreLeadingWhiteSpace",True) \
    .option("multiline", True)\
    .csv("gs://big-data-final/raw-chicago-communities/chi_communities.csv",inferSchema=True, header=True )

chi_communities.show()
```

```python
chi_communities.write.parquet('gs://big-data-final/clean-parquet/chi_communities.parquet')
```

```python
par_df = spark.read.parquet("gs://big-data-final/clean-parquet/chi_communities.parquet")

par_df.show()
```

```python
#turn chicago communities csv into parquet

covid = spark.read \
    .option("quote", "\"")  \
    .option("escape", "\"") \
    .option("ignoreLeadingWhiteSpace",True) \
    .option("multiline", True)\
    .csv("gs://big-data-final/raw-covid/covid.csv",inferSchema=True, header=True )

covid.show()
```

```python
from pyspark.sql import functions as F

renamed_df = covid.select([F.col(col).alias(col.replace(' ', '_')) for col in covid.columns])

renamed_df.show()
```

```python
renamed_df.write.parquet('gs://big-data-final/clean-parquet/covid.parquet')
```
